```python
numDizi1 = np.linspace(0,10,20)
```

```python
numDizi2 = numDizi1 ** 2
```

```python
(myFig,myAxes) = plt.subplots()
myAxes.plot(numDizi1,numDizi2,"y")
myAxes.plot(numDizi2,numDizi1,"g")
```

```python
# renk kodları , alpha ilre transparanlık verilir
```

```python
(myFig,myAxes) = plt.subplots()
myAxes.plot(numDizi1,numDizi2,color = "#3A95A8", alpha = 0.3)
myAxes.plot(numDizi2,numDizi1,color = "#C96F23")
```

```python
#cizgi kalınlıgı, stili linewidth linestyle = "--, :, -."
#marker,markerfacecolor,markersize
```

```python
(myFig,myAxes) = plt.subplots()
myAxes.plot(numDizi1,numDizi1 + 2,color = "b",linewidth = 5.0, linestyle = "-.")
myAxes.plot(numDizi1,numDizi1 + 3,color = "y",linestyle = ":")

myAxes.plot(numDizi1,numDizi1 + 4,color = "b",linestyle = "--")

myAxes.plot(numDizi1,numDizi1 + 7,color = "#000",linestyle = "--", marker = "o", markersize = 4, markerfacecolor = "r")
```
